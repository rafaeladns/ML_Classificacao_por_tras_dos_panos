## 1.2 - Importando os dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/Customer-Churn.csv')
```

```python
dados.shape
```

```python
dados.head()
```

```python
# Dicionário que mapeia os valores 'Sim' para 1 e 'Nao' para 2
traducao_dic = {'Sim': 1, 'Nao': 2}

# Lista com os nomes das colunas do DataFrame que queremos modificar
colunas_para_modificar = ['Conjuge', 'Dependentes', 'TelefoneFixo', 'PagamentoOnline', 'Churn']

# Faz uma cópia do DataFrame original chamado 'dados' para não alterar o original diretamente
dadosmodificados = dados.copy()

# Aplica a substituição de valores nas colunas indicadas, trocando 'Sim' por 1 e 'Nao' por 2
dadosmodificados[colunas_para_modificar] = dadosmodificados[colunas_para_modificar].replace(traducao_dic)

# Mostra as primeiras linhas do DataFrame modificado para verificar se a substituição funcionou
dadosmodificados.head()
```
